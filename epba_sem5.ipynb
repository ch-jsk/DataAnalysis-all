```python
import pandas as pd
import numpy as np
# read datasdet using pandas
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
df.describe()
```

```python
# information about the dataset
df.info()
df.nunique()
```

```python
df["No_acc"].fillna("No No_acc", inplace = True)

df.isnull().sum()
```

```python
mode = df['RegistrationNo'].mode().values[0]
df['RegistrationNo']= df['RegistrationNo'].replace(np.nan, mode)

df.isnull().sum()
```

```python
df = df.dropna(axis = 0, how ='any')

print(df.isnull().sum())
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
# create an instance of LabelEncoder
le = LabelEncoder()

# fit and transform the "Senior Management"
# column with LabelEncoder
df['Occupation'] = le.fit_transform\
					(df['Occupation'])
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.histplot(x='No_acc', data=df, )
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot( x="No_acc", y='RegistrationNo', data=df, )
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot( x="RegistrationNo", y='No_acc', data=df)

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```
