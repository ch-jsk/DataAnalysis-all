```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel("tABLE1.xlsx")
```

```python
df = data.drop(['S2/S3\nmg/g'],axis =1) # removing s2/s3 ratio
```

```python
df = df.rename(columns={'S1\nmg/g':'S1(mg/g)','S2\nmg/g':'S2(mg/g)','S3\nmg/g':'S3(mg/g)','HI\nmg/g':'HI(mg/g)','OI\nmg/g':'OI(mg/g)','PI\nmg/g':'PI(mg/g)','GP\nmg/g':'GP(mg/g)','T max (0C)' : 'T Max(°C)'})
```

```python
df.describe()
```

#Linear Regression


#S1

* Highly dependent variable PI(mg/g), GP(mg/g) , TOC(wt%) and S2(mg/g)

```python
y = df['S1(mg/g)']
x1 = df['HI(mg/g)']
x2 = df['OI(mg/g)']
x3 = df['PI(mg/g)']
x4 = df['GP(mg/g)']
x5 = df['T Max(°C)']
x6 = df['S2(mg/g)']
x7 = df['S3(mg/g)']
x8 = df['TOC (Wt%)']

fig, axs = plt.subplots(2, 4, figsize=(15, 8))
axs = axs.ravel()

# Plot linear regression for each pair of variables
for i, (xi, xlabel) in enumerate(zip([x1, x2, x3, x4, x5, x6, x7, x8], ['HI(mg/g)', 'OI(mg/g)', 'PI(mg/g)', 'GP(mg/g)', 'T Max(°C)', 'S2(mg/g)', 'S3(mg/g)', 'TOC (Wt%)'])):
    sns.regplot(x=xi, y=y, ax=axs[i])
    axs[i].set_xlabel(xlabel)
    axs[i].set_ylabel('S1(mg/g)')

plt.tight_layout()
plt.show()
```

#Correlation

```python
all_corr = df.corr().round(2)

plot = sns.heatmap(all_corr,annot = True)
plt.title('All Coalfield')
```
